```python
# 현대오일뱅크 직영주유소 게시가와 인근 경쟁주유소 간의 가격 차이 추출

def sp_gap(df):
    df2=pd.DataFrame(index=range(0,1), columns=['name', 'date', 'G_sp(won)', 'D_sp(won)','G_gap(won)', 'D_gap(won)','sales(BD)'])
    df2.loc[0, 'name']=df.loc[0,'name']
    df2.loc[0, 'date']=df.loc[0,'date']
    df2.loc[0, 'G_sp(won)']=df.loc[0, 'G']
    df2.loc[0, 'D_sp(won)']=df.loc[0, 'D']
    df2.loc[0, 'G_gap(won)']=df.loc[0,'G']-df["G"].mean()
    df2.loc[0, 'D_gap(won)']=df.loc[0,'D']-df["D"].mean()
    df2.loc[0, 'sales(BD)']=3366/158.984
    
    return df2.head()
```

```python
# 데이터프레임 Column명 전처리

def near_gas(df):
    
    df = pd.DataFrame(df)
    df.columns=[ "name", "address", "date", "brand", "do_brand", "self", "G", "D", "lat", "lon"]
    
    df['g_int']=pd.to_numeric(df['G'])
    df['d_int']=pd.to_numeric(df['D'])
    
    return sp_gap(df)
```

```python
# '주유소 이름' 및 '반경 km 수' 입력 시, 해당 주유소 및 인근 주유소 게시가 정보를 불러오는 함수

from haversine import haversine

def check_distance(df, name, standard =3):
    
    if name not in list(df['name']):
        raise ValueError ('없음')
        
    data = np.array(df)
    lon=np.array(df['lon'])
    lat=np.array(df['lat'])
    idx_a=list(df['name']).index(name)
    loc_a=np.array([lon[idx_a], lat[idx_a]])
    
    list_iter = list(range(len(data)))
    list_iter.remove(idx_a)
    
    result_idx=list()
    result_idx.append(idx_a)
    
    for i in list_iter:
        loc_b = np.array([lon[i], lat[i]])
        if haversine(loc_a, loc_b) < standard :
            result_idx.append(i)
            
    arr=data[result_idx]
    return near_gas(arr)
```

```python
#하나의 데이터프레임의 '현대오일뱅크 전 직졍주유소' 게시가 Gap 정보 입력

def check_group2(df, names, standard=3):
    
    final=list()
    
    for name in names:
        #temporary.. t 파일
        t = check_distance(df, name, standard)
        final.append(t)
        
    return pd.concat(final)
```
